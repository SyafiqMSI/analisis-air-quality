## Visualization & Explanatory Analysis

#### Pertanyaan 1: Bagaimana tren polusi udara pada musim hujan dibandingkan dengan musim kemarau?

```python
def musim(row):
    if row['month'] in [11, 12, 1, 2, 3]: 
        return 'Hujan'
    else: 
        return 'Kemarau'

air_quality_data = air_quality_data.copy()

air_quality_data['season'] = air_quality_data.apply(musim, axis=1)

avg_pm25_by_season = air_quality_data.groupby('season')['PM2.5'].mean()
avg_pm10_by_season = air_quality_data.groupby('season')['PM10'].mean()

seasons = ['Hujan', 'Kemarau']
pm25_means = [avg_pm25_by_season['Hujan'], avg_pm25_by_season['Kemarau']]
pm10_means = [avg_pm10_by_season['Hujan'], avg_pm10_by_season['Kemarau']]

plt.bar(seasons, pm25_means, color='blue', alpha=0.6, label='PM2.5')
plt.bar(seasons, pm10_means, color='orange', alpha=0.6, label='PM10', bottom=pm25_means)
plt.title('Rata-rata Kualitas Udara antara Musim Hujan dan Musim Kemarau')
plt.xlabel('Musim')
plt.ylabel('Rata-rata Kualitas Udara')
plt.legend()
plt.show()

```

## Analisis Lanjutan (Opsional)

#### Clustering tingkat polusi

```python
def categorize_pollution(row):
    if row['PM2.5'] < 12 and row['PM10'] < 20:
        return 'Rendah'
    elif (12 <= row['PM2.5'] < 35) and (20 <= row['PM10'] < 50):
        return 'Sedang'
    else:
        return 'Tinggi'

all_data_clean['Pollution_Level'] = all_data_clean.apply(categorize_pollution, axis=1)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PM2.5', y='PM10', hue='Pollution_Level', data=all_data_clean, palette='viridis')
plt.title('Clustering PM2.5 dan PM10')
plt.xlabel('PM2.5 (g/m)')
plt.ylabel('PM10 (g/m)')
plt.legend(title='Tingkat Polusi')
plt.show()
```

#### Clustering wind speed

```python
def categorize_wind_speed(row):
    if row['WSPM'] < 3:
        return 'Rendah'
    elif 3 <= row['WSPM'] < 7:
        return 'Sedang'
    else:
        return 'Tinggi'

all_data_clean['Wind_Speed_Level'] = all_data_clean.apply(categorize_wind_speed, axis=1)

plt.figure(figsize=(10, 6))
sns.countplot(x='Wind_Speed_Level', data=all_data_clean)
plt.title('Clustering Kecepatan Angin')
plt.xlabel('Tingkat Kecepatan Angin')
plt.ylabel('Frekuensi')
plt.show()
```

#### Hubungan Polusi dengan kecepatan angin

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(y='PM2.5', x='WSPM', hue='PM10',style='Wind_Speed_Level', data=all_data_clean, palette='viridis')
plt.title('Hubungan antara Polusi dan Kecepatan Angin')
plt.xlabel('PM2.5 & PM10 (g/m)')
plt.ylabel('Kecepatan Angin (m/s)')
plt.legend(title='Tingkat Kecepatan Angin')
plt.show()
```

#### Linear Regression

```python
X = all_data_clean[['TEMP', 'DEWP', 'WSPM']]
y_pm25 = all_data_clean['PM2.5']
y_pm10 = all_data_clean['PM10']

def linear_regression(X, y):
    X = np.column_stack((np.ones(X.shape[0]), X)) 

    y_mean = np.mean(y)
    X_mean = np.mean(X, axis=0)

    b = np.zeros(X.shape[1])  
    for j in range(1, X.shape[1]): 
        numerator = np.sum((X[:, j] - X_mean[j]) * (y - y_mean))
        denominator = np.sum((X[:, j] - X_mean[j])**2)
        b[j] = numerator / denominator
    b[0] = y_mean - np.sum(b[1:] * X_mean[1:])

    return b

coefficients_pm25 = linear_regression(X.to_numpy(), y_pm25.to_numpy())
coefficients_pm10 = linear_regression(X.to_numpy(), y_pm10.to_numpy())

print("Koefisien untuk PM2.5:", coefficients_pm25)
print("Koefisien untuk PM10:", coefficients_pm10)
```

**Insight:**
- Mengkategorikan tingkat polusi berdasarkan nilai PM2.5 dan PM10 menjadi tiga: Rendah, Sedang, dan Tinggi pada satauan (g/m)
- Mengkategorikan kecepatan angin menjadi tiga tingkat: Rendah, Sedang, dan Tinggi.
- Mengkategorikan untuk menunjukkan hubungan antara polusi (PM2.5 dan PM10) dan kecepatan angin 
- Regresi linier untuk mengukur hubungan antara variabel dengan tingkat polusi (PM2.5 dan PM10).
